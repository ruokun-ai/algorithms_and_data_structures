  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[1, n]`. Return a list of numbers that are missing from the range `[1, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [1, 1, 3]`

  Output: [2]

  ### Example 2

  Input: `lst = [5,3,2,4,1]`

  Output: [-1]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 1, 1, 2,10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: List) ->[int]:
    n = len(nums)
    nums_set = set(nums)

    missing = []

    for i in range(1, n + 1):
        if i not in nums_set:
            missing.append(i)

    if len(missing) == 0:
        return [-1]

    return missing


# Example #1
lst = [1, 1, 3]
print("Example 1 \nInput", lst, "\nOutput:", missing_num(lst), "\n")

# Example #2
lst = [5, 3, 2, 4, 1]
print("Example 2 \nInput", lst, "\nOutput:", missing_num(lst), "\n")

# Example #3
lst = [6, 8, 2, 3, 5, 7, 1, 1, 2, 10]
print("Example 3 \nInput", lst, "\nOutput:", missing_num(lst))
```

```python
# Code for Question 1
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
    
    return -1
```
